```julia
using Plots
```

```julia
N = 50;                            # number of points in each direction
x_start, x_end = -2.0, 2.0        # boundaries in the x-direction
y_start, y_end = -1.0, 1.0        # boundaries in the y-direction
x = range(x_start, stop=x_end, length=N);      # 1D-array with x-coordinates
y = range(y_start, stop=y_end, length=N);      # 1D-array with y-coordinates

println(size(x))
println(size(y))

meshgrid(xs, ys) = [xs[i] for i in 1:length(xs), 
        j in length(ys)], [ys[j] for i in 1:length(xs), 
        j in 1:length(ys)];                   # defining a meshgrid function

X, Y = meshgrid(x, y);
println(size(X), size(Y))
```

```julia
# plot grid points
width = 10;
height = (y_end - y_start) / (x_end - x_start) * width;
scatter(X, Y, color=:red, marker=:circle, label=:none, markersize=2)
```

```julia
strength_source = 5.0;                  # source strength
x_source, y_source = -1.0, 0.0;         # location of the source

# compute the veloicty field on the mesh grid
u_source = (strength_source / (2 *pi)) .*
            (X .- x_source) ./ ((X .- x_source) .^2 + (Y .- y_source) .^2);
v_source = (strength_source / (2 *pi)) .*
            (Y .- y_source) ./ ((X .- x_source) .^2 + (Y .- y_source) .^2);
```

```julia
#plot the streamlines

streamplot(X, Y, u_source, v_source,
                  density=2, linewidth=1, arrowsize=2, arrowstyle="->")
scatter(x_source, y_source,
               color="#CD2305", s=80, marker="o");
```
